## Task 8 - Web scraping

#### Task 8.1
How much do the top companies cost?

Analyze this link: [Companies Market Cap - Top Companies](https://companiesmarketcap.com/)

Create a list of the names of the top 10 companies by market capitalization on this page.

```python
import requests
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://companiesmarketcap.com/'
```

```python
response = requests.get(url)
response
```

```python
soup = bs(response.text)
type(soup)
```

```python
top = soup.find(class_ =  'company-name').text
top
```

```python
names = []
for i in soup.find_all(class_ = 'company-name'):
    names.append(i.text)
names
```

```python
market_cap = []



for i in soup.find_all(class_ = 'td-right'):
    if i.text.startswith('$'):
        market_cap.append(i.text)
market_cap = market_cap[::2]
market_cap
```

```python
share = []


for i in soup.find_all(class_ = 'td-right'):
    if i.text.startswith('$'):
        share.append(i.text)

share = share[1::2]
share
```

```python
country = []

for i in soup.find_all(class_ = 'responsive-hidden'):
    country.append(i.text)
country = country[3:]
country
```

```python
data = {

    'Companies'  : names ,
    'Market_cap' : market_cap,
    'Share_price': share,
    'Country'    : country
}
```

```python
df = pd.DataFrame(data)
df
```
